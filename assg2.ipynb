```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(10)#messi for a reason
```

```python
n_s = [1, 5, 10]
sam = []
for i in range(1, 1000):
    x= np.random.chisquare(3, 1)
    mu = np.mean(x)
    sam.append(mu)
    
plt.hist(sam, bins = 30, density=True,)
x=np.linspace(0, 10, 1000)
pdf = stats.norm.pdf(x, 3, np.sqrt(6/1))
plt.plot(x, pdf, 'k--', label = 'Gaussian')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
url = "http://www.iith.ac.in/~shantanud/test.dat"
data = np.loadtxt(url)
luminosity = data[:, 0]
redshift = data[:, 1]
plt.figure(figsize=(8, 6))
plt.scatter(redshift, luminosity, marker='o', alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Redshift (log scale)')
plt.ylabel('Luminosity (log scale)')
plt.title('Luminosity vs Redshift')
plt.show()
spearman_corr, spearman_p_value = stats.spearmanr(redshift, luminosity)
pearson_corr, pearson_p_value = stats.pearsonr(redshift, luminosity)
kendall_corr, kendall_p_value = stats.kendalltau(redshift, luminosity)
print(f"Spearman correlation coefficient: {spearman_corr:.4f}, p-value: {spearman_p_value:.4f}")
print(f"Pearson correlation coefficient: {pearson_corr:.4f}, p-value: {pearson_p_value:.4f}")
print(f"Kendall tau correlation coefficient: {kendall_corr:.4f}, p-value: {kendall_p_value:.4f}")
```

```python
a1 = np.random.normal(0, 1, 1000)
a2 = np.random.normal(0, 1, 1000)
pearson_corr, pearson_pval = stats.pearsonr(a1, a2)
dof = len(a1)-2
t_stat = pearson_corr*np.sqrt(dof/(1-pearson_corr**2))
t_pval = 2*(1-stats.t.cdf(np.abs(pearson_corr)/np.sqrt((1-pearson_corr**2)/ dof), dof))
print(f"Pearson correlation coefficient: {pearson_corr:.4f}")
print(f"P-value (Pearson correlation): {pearson_pval:.4f}")
print(f"\nT-statistic: {t_stat:.4f}")
print(f"P-value (Student-t distribution): {t_pval:.4f}")
if np.isclose(pearson_pval, t_pval, rtol=1e-5):
    print("The p-value agrees with that calculated using the Student-t distribution.")
else:
    print("The p-value does not agree with that calculated.")
```
